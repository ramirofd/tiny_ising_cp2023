```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data into a pandas dataframe
df = pd.read_csv('../results/size_comparison_32_1024_local.csv')

# Use seaborn to create a line plot
sns.scatterplot(x='L', y='time', data=df)

# Set the plot title and axis labels
plt.title('Time vs Grid Size')
plt.xlabel('Grid Size')
plt.ylabel('Time')
plt.xticks(df.L.unique(), rotation=90)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data into a pandas dataframe
df = pd.read_csv('../results/optimization_flags_local.csv')

# Use seaborn to create a line plot
print(df.columns)
ax = sns.barplot(x='flags', y='metric', hue='compiler', data=df)
# Add labels with the value of each bar
for rect in ax.patches:
    # Get the height and width of the bar
    height = rect.get_height()
    width = rect.get_width()

    # Get the x and y coordinates of the bar
    x = rect.get_x()
    y = rect.get_y()

    # Add a label with the value of the bar
    ax.text(x + width/2, y + height/2, '{:.2f}'.format(height), ha='center', va='center', color='white', rotation=90)

# Set the plot title and axis labels
plt.title('Metric vs Flags for GCC/CLang')
plt.xlabel('Flags')
plt.ylabel('Metric [p/s]')
# plt.xticks(df.L.unique(), rotation=90)

# Show the plot
plt.show()
```
